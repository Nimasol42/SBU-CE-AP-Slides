    <li>آیا یک <strong>InputStream</strong> جریانی است که برنامه شما داده‌ها را از آن می‌خواند؟</li>
    <li>در این درس، تفاوت دقیق این مفاهیم آشکار می‌شوند.</li>

  <ul>
    <li>اگر برنامه بخواهد داده‌ای را از یک منبع بخواند، به یک کلاس <strong>Reader</strong> یا <strong>InputStream</strong> نیاز دارد.</li>
    <li>اگر برنامه بخواهد داده‌ای را در یک مقصد بنویسد، به یک کلاس <strong>Writer</strong> یا <strong>OutputStream</strong> نیاز دارد.</li>
    <li>یک <strong>InputStream</strong> یا <strong>Reader</strong> به یک منبع داده‌ای متصل می‌شود.</li>
    <li>یک <strong>Writer</strong> یا <strong>OutputStream</strong> به یک مقصد داده‌ای متصل می‌شود.</li>

<div class="card" dir="rtl">
  <h3>کلاس File</h3>
  <ul>
    <li>با استفاده از کلاس <strong>File</strong> در <code>Java IO API</code> می‌توانیم به سیستم فایل کامپیوتر دسترسی پیدا کنیم.</li>
    <li>کلاس <strong>File</strong> به شما امکان می‌دهد:</li>
    <ul>
      <li>بررسی کنید آیا یک فایل یا دایرکتوری (فهرست) وجود دارد؟</li>
      <li>فایل یا مسیر فایل را در قالب نام فایل که یک رشته است، ایجاد کنید.</li>
      <li>اطلاعات مربوط به فایل مانند وجود فایل (دایرکتوری)، قابلیت خواندن یا نوشتن آن، طول آن، مسیر آن، زمان تغییر، نام آن، پنهان بودن آن را کسب نمایید.</li>
      <li>لیستی از فایل‌ها درون یک دایرکتوری را بخوانید.</li>
      <li>همچنین می‌توانیم فایل را پاک کنیم یا نامش را تغییر دهیم.</li>
      <li>در اسلایدهای بعدی برخی از متدهای کلاس <strong>File</strong> که بر روی اشیاء حاصل از این کلاس <strong>File</strong> اعمال هستند، نشان داده شده‌اند.</li>
    </ul>
    <li>توجه کنید که کلاس <strong>File</strong> به شما امکان خواندن از یا نوشتن در فایل‌ها را نمی‌دهد.</li>
    <li>برای خواندن/نوشتن باید از کلاس‌های <strong>FileInputStream</strong> و <strong>FileOutputStream</strong> و <strong>RandomAccessFile</strong> استفاده کنید.</li>
  </ul>

  <p> - کسب اطلاعات از خصوصیات فایل یا دستکاری آن</p>
  <div style="  display: block; margin: auto; width: 70%;" class="image">
    <img src="img/java_io_file.png" alt="Java io file" />
  </div>

  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">ایجاد یک شیء از کلاس <code>java.io.File</code></h4>
  <ul>
    <li>قبل از آنکه بخواهید با یک سیستم فایل کار کنید یا از امکانات کلاس فایل استفاده کنید باید ابتدا یک نمونه (شیء) از کلاس <strong>File</strong> ایجاد کنید.
</li>
    <li>برای مثال:</li>
    <p dir="ltr"><code>file = new File("e:\\data\\input-file.txt");</code></p>
    <li>این یک نمونه از سازنده‌های کلاس <strong>File</strong> است. این کلاس سازنده‌های مشابهی نیز دارد که بسته به کاربردتان می‌توانید از هریک استفاده کنید.</li>
    <li> شیء <strong>File</strong> مسیر <code>"c:\data\input-file.txt"</code> را در قالب یک رشته نگهداری می‌کند و از این پس قادر است عملیاتی را بر روی این مسیر (فایل یا دایرکتوری) انجام دهد. </li>
  </ul>
  
  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">بررسی وجود یه فایل</h4>
  <ul>
    <li>پس از ایجاد یک شیء از کلاس فایل می‌توانید بررسی کنید آیا فایل مشخص‌شده در رشته حاوی مسیر در شیء، وجود خارجی دارد یا خیر.</li>
    <li>سازنده کلاس <strong>File</strong> حتی اگر چنین فایلی در مسیر مشخص‌شده وجود نداشته باشد، دچار اشکال نمی‌شود.</li>
    <li>اگر چنین فایلی وجود نداشته باشد، می‌توانید آن را ایجاد کنید.</li>
    <li>مثال زیر را ببینید:</li>
    <p dir="ltr"><code>File file = new File("c:\\data\\input-file.txt");</code></p>
    <p dir="ltr"><code>boolean fileExists = file.exists();</code></p>
    <li>این متد اگر چنین مسیری وجود داشته باشد، <strong>true</strong> بر می‌گرداند. اگر فایل مربوطه در مسیر مشخص‌شده وجود نداشته باشد، با دستور زیر آن را ایجاد کنید:</li>
    <p dir="ltr"><code>file.createNewFile();</code></p>
    <p>با فرض آنکه دایرکتوری <code dir="ltr">c:\users\AP</code> از قبل موجود باشد، کد بالا یک زیردایرکتوری به نام <strong>Java</strong> می‌سازد.</p>
    <li>برای ایجاد دایرکتوری‌ها از دستورات <code dir="ltr">mkdir()</code> و <code dir="ltr">mkdirs()</code> در ادامه استفاده کنید.</li>
  </ul>

  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">ایجاد یک دایرکتوری اگر موجود نباشد</h4>
  <ul>
    <li>با استفاده از کلاس <strong>File</strong> می‌توانید درصورت عدم وجود دایرکتوری مربوطه، دایرکتوری ایجاد کنید.</li>
    <li>کلاس <strong>File</strong> دو متد <code dir="ltr">mkdir()</code> و <code dir="ltr">mkdirs()</code> دارد که برای شما طبق مسیری که مشخص کرده‌اید، دایرکتوری می‌سازند.</li>
    <li>متد <code dir="ltr">mkdir()</code> در صورت عدم وجود دایرکتوری، یک دایرکتوری جدید می‌سازد.</li>
    <li>مثال:</li>
    <p dir="ltr"><code>File file = new File("c:\\users\\AP\\Java");</code></p>
    <p dir="ltr"><code>boolean dirCreated = file.mkdir();</code></p>
    <p>با فرض آنکه دایرکتوری <code dir="ltr">c:\users\AP</code> از قبل موجود باشد، کد بالا یک زیردایرکتوری به نام <strong>Java</strong> می‌سازد.</p>
    <li>متد <code dir="ltr">mkdir()</code> مقدار <strong>true</strong> برمی‌گرداند اگر دایرکتوری ایجاد شود و در غیراینصورت <strong>false</strong> برمی‌گرداند.</li>
    <li>اگر بخواهید کل دایرکتوری را بسازید (اگر مسیر <code>C:\users\AP</code> از قبل موجود نباشد) از متد <code dir="ltr">mkdirs()</code> استفاده کنید.</li>
    <p dir="ltr"><code>File file = new File("c:\\Users\\AP\\newdir");</code></p>
    <p dir="ltr"><code>boolean dirCreated = file.mkdirs();</code></p>
  </ul>

  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">اندازه فایل</h4>
  <ul>
    <li>با متد <code dir="ltr">length()</code> می‌توانید از اندازه فایل دلخواه برحسب بایت آگاه شوید.</li>
    <li>مثال:</li>
    <p dir="ltr"><code>File file = new File("c:\\data\\input-file.txt");</code></p>
    <p dir="ltr"><code>long length = file.length();</code></p>    
  </ul>

  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">تغییر نام یا جابجایی فایل</h4>
  <ul>
    <li>برای تغییر نام یا جابجایی یک فایل از متد <code dir="ltr">renameTo()</code> استفاده کنید.</li>   
    <li>مثال:</li>   
    <p dir="ltr"><code>File file = new File("c:\\data\\input-file.txt");</code></p>   
    <p dir="ltr"><code>boolean success = file.renameTo(new File("c:\\data\\new-file.txt"));</code></p>   
    <li>با این متد همچنین می‌توانید یک فایل را به یک دایرکتوری جدید منتقل کنید.</li>   
    <li>لزومی ندارد که نام فایل جدید در همان مسیر فایل قبلی باشد.</li>   
    <li>اگر عمل تغییر نام یا جابجایی موفقیت‌آمیز بود، مقدار <strong>true</strong> برگردانده خواهد شد و در غیراینصورت <strong>false</strong> برمی‌گردد (مثلاً اگر فایل باز باشد).</li>   
  </ul>

  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">پاک کردن فایل</h4>
  <ul>
    <li>برای حذف فایل از متد <code dir="ltr">delete()</code> استفاده می‌کنیم.</li>
    <li>مثال:</li>
    <p dir="ltr"><code>File file = new File("c:\\data\\input-file.txt");</code></p>
    <p dir="ltr"><code>boolean success = file.delete();</code></p>
    <li>بسته به موفقیت‌آمیز بودن یا نبودن این عملیات به ترتیب مقادیر <strong>true</strong> یا <strong>false</strong> برمی‌گردد.</li>
  </ul>

  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">خواندن لیستی از فایل‌ها در دایرکتوری</h4>
  <ul>
    <li>برای دانستن لیست پوشه‌ها و فایل‌های موجود در یک دایرکتوری از متدهای <code dir="ltr">list()</code> یا <code dir="ltr">listFiles()</code> استفاده می‌کنیم.</li>
    <li>خروجی این متدها آرایه‌ای از نوع <code>String</code> می‌باشد.</li>
    <li>مسیری که می‌خواهیم محتویات آن را بازیابی کنیم، باید از جنس دایرکتوری باشد.</li>
    <li>مثال:</li>
    <ul dir="ltr">
      <li><code>File file = new File("c:\\data");</code></li>
      <li><code>String[] fileNames = file.list();</code></li>
    </ul>
    <ul>
      <li>محتویات دایرکتوری مشخص شده را در یک آرایه از رشته‌ها ذخیره‌سازی می‌کند.</li>
    </ul>
    <ul dir="ltr">
      <li><code>File[] files = file.listFiles();</code></li>
    </ul>
    </ul>
    <li>محتویات دایرکتوری مشخص شده را در یک آرایه از فایل‌ها (از نوع شیء کلاس <strong>File</strong>) ذخیره‌سازی می‌کند.</li>
  </ul>


  <hr class="divider" />

  <h4 style="color:red;  text-align: center;">مثالی از کلاس فایل</h4>


```java
public class TestFileClass {
    public static void main(String[] args) throws IOException{
    java.io.File file = new java.io.File(" C:\\Test\\test.txt ");

    file.createNewFile();
    System.out.println("Does it exist? " + file.exists());
    System.out.println("The file has " + file.length() + " bytes");
    System.out.println("Can it be read? " + file.canRead());
    System.out.println("Can it be written? " + file.canWrite());
    System.out.println("Is it a directory? " + file.isDirectory());
    System.out.println("Is it a file? " + file.isFile());
    System.out.println("Is it absolute? " + file.isAbsolute());
    System.out.println("Is it hidden? " + file.isHidden());
    System.out.println("Absolute path is " +  file.getAbsolutePath());
    System.out.println("Last modified on " +  new java.util.Date(file.lastModified()));
    }
}
```
