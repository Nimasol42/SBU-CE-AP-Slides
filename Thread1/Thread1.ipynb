  <p>
    قبل از شروع درس امروز دوست داشتم یه دور مفاهیم رو با هم مرور کنیم که جلوتر به مشکل نخوریم.

    به هر برنامه‌ای که در زمان اجرا داریم، میگیم <strong>process</strong>.  
    نکته مهم اینه که مسئول اجرای processها همون واحد پردازشی کامپیوتر، یعنی <strong>CPU</strong> هست.  
  </p>
  <p>
    در واقع هر برنامه به یه سری دستور به زبان‌هایی مثل <em>assembly</em> تبدیل میشه.  
    مثلا فرض کنید برنامه ما ۵۰۰ خط دستور داره. اگه CPU تک‌هسته‌ای داشته باشیم، در هر لحظه فقط یه دستور می‌تونه اجرا بشه.  
  </p>
  <p>

  </p>
  <p>

    یه مدت کوتاه روی یکی اجرا می‌کنه، بعد میره سراغ یکی دیگه.  
    در واقع CPU بین همه processها مشترکه، اما <strong>در هر لحظه فقط یک دستور از یک process رو اجرا می‌کنه</strong>.
  </p>

  <h2>واژه‌نامه خودمونی</h2>
  <ul>
    <li>
      <strong>فرآیند یا برنامه(Process)</strong>  
      یه برنامه در حال اجراست. مثلا وقتی موزیک پلیر رو باز می‌کنی، اون تبدیل به یه پراسس میشه که CPU باید دستوراتشو اجرا کنه.
    </li>
    <li>
      <strong>نخ (Thread)</strong>  
      یه مسیر سبک‌تر و کوچیک‌تر از پردازشه. انگار توی همون برنامه چندتا کارگر جدا داشته باشی که هر کدوم یه بخش از کار رو انجام بدن.
    </li>
    <li>
      <strong>زیر‌برنامه یا زیرفرآیند (Subprocess)</strong>  
       یعنی یه پراسس جدید که توسط یه پراسس دیگه ساخته میشه. مثلا وقتی برنامه‌ت یه برنامه دیگه رو صدا بزنه.
    </li>
    <li>
      <strong>(Context Switch)</strong>  
      وقتی CPU از اجرای یه پراسس/ترد دست می‌کشه و میره سراغ یه پراسس/ترد دیگه. مثل وقتی که معلم از سوال یکی ول می‌کنه میره جواب نفر دیگه رو بده.
    </li>
    <li>
      <strong>(Multitasking)</strong>  
      یعنی انجام چندتا کار با هم. در عمل CPU خیلی سریع بین کارها جابه‌جا میشه، جوری که به چشم ما میاد همه‌چی با هم در حال اجراست.
    </li>
  </ul>
</div>

  <h2>مقدمه: مالتی‌تردینگ در جاوا</h2>
  <p>
    مالتی‌تردینگ (<strong>Multithreading</strong>) فرآیند اجرای همزمان چندین نخ (Thread) است. 
    نخ‌ها در واقع زیرفرآیندهای سبک هستند و کوچک‌ترین واحد پردازشی در یک برنامه محسوب می‌شوند. 
    هدف اصلی مالتی‌تردینگ، <strong>افزایش کارایی</strong> و <strong>انجام چند کار به طور همزمان</strong> با استفاده از منابع مشترک است.
  </p>

 <h4 style="color: #dc2626;">چرا از مالتی‌تردینگ استفاده می‌کنیم؟</h4>
  <ul>
    <li>به جای چندین فرآیند سنگین، از نخ‌ها استفاده می‌کنیم چون نخ‌ها حافظه مشترک دارند و سبک‌تر هستند.</li>
    <li>جابجایی بین نخ‌ها (Context Switching) سریع‌تر از جابجایی بین فرآیندها است.</li>
    <li>باعث <strong>صرفه‌جویی در حافظه</strong> و <strong>کاهش زمان اجرا</strong> می‌شود.</li>
    <li>کاربردهای گسترده در برنامه‌های گرافیکی، بازی‌ها و انیمیشن‌ها دارد.</li>
  </ul>

  <h4 style="color: #dc2626;">مزایای مالتی‌تردینگ در جاوا</h4>
  <ol>
    <li>برنامه هنگام اجرای یک نخ، نخ‌های دیگر را بلاک نمی‌کند.</li>
    <li>امکان انجام همزمان چند عملیات را فراهم می‌کند، در نتیجه زمان کمتری مصرف می‌شود.</li>
    <li>اگر یک نخ دچار خطا شود، نخ‌های دیگر مستقل عمل کرده و متوقف نمی‌شوند.</li>
  </ol>
</div>

<div class="card" dir="rtl">
  <h2>مالتی‌تسکینگ (Multitasking)</h2>
  <p>
    مالتی‌تسکینگ یعنی اجرای چند کار به صورت همزمان.  
    ما از مالتی‌تسکینگ استفاده می‌کنیم تا CPU رو به بهترین شکل ممکن به کار بگیریم.
  </p>

  <h4 style="color: #dc2626;">انواع مالتی‌تسکینگ</h4>
  <ol>
    <li>
      <strong>مالتی‌تسکینگ مبتنی بر پراسس (Process-based Multitasking یا Multiprocessing)</strong>
      <ul>
        <li>هر پراسس آدرس جداگانه در حافظه داره (یعنی هر برنامه فضای حافظه مخصوص به خودش رو می‌گیره).</li>
        <li>پراسس‌ها سنگین (Heavyweight) هستن.</li>
        <li>هزینه ارتباط (Communication) بین پراسس‌ها بالاست.</li>
        <li>جابجایی از یک پراسس به پراسس دیگه (Context Switch) زمان‌بره چون باید رجیسترها، مپ‌های حافظه و لیست‌ها ذخیره و بارگذاری بشن.</li>
      </ul>
    </li>
    <li>
      <strong>مالتی‌تسکینگ مبتنی بر ترد (Thread-based Multitasking یا Multithreading)</strong>
      <ul>
        <li>همه تردها یک فضای آدرس مشترک دارن.</li>
        <li>تردها سبک (Lightweight) هستن.</li>
        <li>هزینه ارتباط بین تردها پایین و سریع‌تره.</li>
      </ul>
    </li>
  </ol>
</div>

<div class="card" dir="rtl">
  <h2>ترد (Thread) در جاوا</h2>
  <p>
    ترد در واقع یک <strong>زیرفرآیند سبک (Lightweight Subprocess)</strong> و کوچک‌ترین واحد پردازشی در سیستم است.  
    هر ترد یک مسیر جداگانه برای اجرا دارد.  
    نخ‌ها مستقل از هم عمل می‌کنند؛ اگر در یکی خطایی رخ بدهد، روی بقیه تأثیر نمی‌گذارد.  
    همه‌ی تردها از یک ناحیه حافظه مشترک استفاده می‌کنند.
  </p>

  <h4 style="color: #dc2626;">ویژگی‌های اصلی ترد</h4>
  <ul>
    <li>یک ترد همیشه داخل یک پراسس اجرا می‌شود.</li>
    <li>بین تردها <strong>context switching</strong> انجام می‌شود (جابجایی سریع CPU بین نخ‌ها).</li>
    <li>هر سیستم‌عامل می‌تواند چندین پراسس داشته باشد و هر پراسس هم می‌تواند شامل چندین ترد باشد.</li>
  </ul>

  <div class = "images-big">
    <img src="img/pic1.png"/>
  </div>
  <p>
    در این تصویر می‌بینیم که سیستم‌عامل (OS) شامل چند پراسس مختلف است (Process 1، Process 2 و Process 3).  
    هر پراسس می‌تواند شامل یک یا چند ترد باشد. برای مثال:
  </p>
  <ul>
    <li>Process 1 شامل سه ترد (t1, t2, t3) است.</li>
    <li>Process 2 شامل دو ترد (t1, t2) است.</li>
    <li>Process 3 فقط یک ترد دارد.</li>
  </ul>

  <p>
    اگر CPU ما <strong>فقط یک هسته (Single Core)</strong> داشته باشد، باز هم امکان اجرای چند ترد وجود دارد.  
    در این حالت CPU به کمک <em>context switching</em> بین تردها جابجا می‌شود.  
    مثلا CPU برای چند میلی‌ثانیه دستورهای t1 را اجرا می‌کند، بعد سراغ t2 می‌رود، بعد t3 و دوباره برمی‌گردد به t1.  
    سرعت این جابجایی آنقدر بالاست که به چشم ما همه‌ی تردها همزمان در حال اجرا به نظر می‌آیند.
  </p>

  <h4 style="color: #dc2626;">یک مثال ساده</h4>
  <p>
    فرض کنید شما در حال گوش دادن به موزیک (یک ترد) هستید، همزمان در مرورگر (ترد دیگر) جزوه می‌خوانید و در پس‌زمینه هم یک دانلود (ترد دیگر) در حال انجام است.  
    اگر CPU تک‌هسته‌ای باشد، هر بار فقط یک دستور از یکی از این تردها اجرا می‌شود.  
    اما به خاطر جابجایی سریع CPU بین این کارها، شما حس می‌کنید همه‌چیز همزمان در حال انجام است.
  </p>
</div>

<div class="card" dir="rtl">
  <h2>کلاس Thread در جاوا</h2>
  <p>
    جاوا برای کار با نخ‌ها (<strong>Threads</strong>) یک کلاس به اسم <code>Thread</code> در اختیار ما قرار داده.  
    این کلاس متدها و سازنده‌های مختلفی داره که به کمکشون می‌تونیم نخ‌ها رو ایجاد و مدیریت کنیم.  
    در واقع <code>Thread</code> از کلاس <code>Object</code> ارث‌بری می‌کنه و اینترفیس <code>Runnable</code> رو هم پیاده‌سازی کرده.
  </p>

  <h4 style="color: #dc2626;">کاربردهای اصلی کلاس Thread</h4>
  <ul>
    <li>ایجاد یک نخ جدید.</li>
    <li>شروع اجرای نخ با استفاده از متد <code>start()</code>.</li>
    <li>اجرای کد داخل نخ با متد <code>run()</code>.</li>
    <li>مدیریت ویژگی‌هایی مثل نام نخ، اولویت نخ و وضعیت اجرا.</li>
  </ul>

  <h4 style="color: #dc2626;">سازنده‌های پرکاربرد Thread</h4>
  <ul>
    <li><code>Thread()</code> : ایجاد یک نخ خالی.</li>
    <li><code>Thread(Runnable target)</code> : ایجاد نخ جدید با استفاده از یک شیء که <code>Runnable</code> پیاده‌سازی کرده.</li>
    <li><code>Thread(String name)</code> : ایجاد نخ با نام مشخص.</li>
    <li><code>Thread(Runnable target, String name)</code> : ایجاد نخ با runnable و نام مشخص.</li>
  </ul>
</div>

  <h2>چرخه زندگی ترد (Thread Life Cycle)</h2>
  <p>
    یک نخ (Thread) در طول عمر خودش می‌تونه در وضعیت‌های مختلفی قرار بگیره.  
    به طور کلی، چرخه‌ی زندگی نخ توسط <strong>JVM</strong> کنترل میشه.  
    طبق تعریف Sun چهار حالت اصلی وجود داره، اما برای درک بهتر معمولا اون رو در پنج حالت توضیح می‌دن.
  </p>

  <div class = "images-big">
    <img src="img/pic2.png"/>
  </div>


  <h4 style="color: #dc2626;">حالت‌های چرخه‌ی نخ</h4>
  <ol>
    <li><strong>New</strong> : وقتی که یک شیء از کلاس <code>Thread</code> ساخته میشه اما هنوز متد <code>start()</code> صدا زده نشده.</li>
    <li><strong>Runnable</strong> : بعد از اجرای متد <code>start()</code> نخ آماده‌ی اجراست اما هنوز CPU بهش تخصیص داده نشده، درواقع هنوز توسط Thread Scheduler انتخاب نشده.</li>
    <li><strong>Running</strong> : وقتی که Thread Scheduler نخ رو انتخاب کنه و در حال اجرا باشه. یعنی CPU بهش اختصاص داده شده.</li>
    <li><strong>Non-Runnable (Blocked/Waiting)</strong> : نخ هنوز زنده‌ست ولی در حال حاضر نمی‌تونه اجرا بشه و CPU از دست نخ خارج شده و CPU آماده برای سرویس دادن به نخ دیگه‌‌ای هستش. (مثلا با متد <code>sleep()</code> یا <code>join()</code> منتظر مونده).</li>
    <li><strong>Terminated (Dead)</strong> : وقتی که اجرای متد <code>run()</code> تموم بشه، نخ به وضعیت خاتمه یافته میره.</li>
  </ol>

<h4 style="color: #dc2626;">توضیح ساده با مثال</h4>
<p>
  فرض کنید یک مرورگر وب (Browser) رو باز می‌کنید.  
  وقتی تب جدید رو باز می‌کنید اما هنوز چیزی لود نشده، مثل حالتیه که نخ تازه ساخته شده (New).  
  به محض اینکه روی یک لینک کلیک می‌کنید و مرورگر آماده‌ی لود کردن صفحه میشه، وارد حالت Runnable میشه.  
  وقتی CPU شروع می‌کنه به اجرای درخواست و پردازش داده‌ها، نخ در حالت Running قرار داره.  
  حالا اگه مرورگر منتظر پاسخ سرور بمونه یا برای لحظاتی به حالت تعلیق بره (مثلا در حال دانلود باشه)، در وضعیت Non-Runnable قرار می‌گیره.  
  و در نهایت وقتی بارگذاری صفحه کامل شد یا تب رو بستید، اون نخ کارش تموم شده و وارد حالت Terminated میشه.
</p>

  <h2>روش‌های ایجاد ترد در جاوا</h2>
  <p>
    در جاوا دو راه اصلی برای ایجاد و اجرای نخ وجود داره:
  </p>

  <h4 style="color: #dc2626;">۱. ایجاد نخ با ارث‌بری از کلاس Thread</h4>
  <p>
    در این روش یک کلاس جدید می‌سازیم که از <code>Thread</code> ارث‌بری می‌کنه و متد <code>run()</code> رو بازنویسی می‌کنیم.  
    برای شروع نخ هم از متد <code>start()</code> استفاده می‌کنیم.
  </p>
</div>

```java
class MyThread extends Thread {
    public void run() {
        System.out.println("Thread is running by extending Thread class...");
    }

    public static void main(String[] args) {
        MyThread t1 = new MyThread();
        t1.start(); 
    }
}

MyThread.main(new String[]{});
```

<link rel="stylesheet" href="style.css">
<div class="card" dir="rtl">
  <h4 style="color: #dc2626;">۲. ایجاد نخ با پیاده‌سازی Runnable Interface</h4>
  <p>
    در این روش یک کلاس می‌سازیم که <code>Runnable</code> رو پیاده‌سازی می‌کنه و متد <code>run()</code> رو تعریف می‌کنیم.  
    بعد باید یک شیء از <code>Thread</code> بسازیم و شیء <code>Runnable</code> رو بهش پاس بدیم.  
  </p>
</div>

```java
class MyRunnable implements Runnable {
    public void run() {
        System.out.println("Thread is running by implementing Runnable interface...");
    }

    public static void main(String[] args) {
        MyRunnable myRunnable = new MyRunnable();
        Thread t1 = new Thread(myRunnable);
        t1.start(); 
    }
}

MyRunnable.main(new String[]{});
```

<link rel="stylesheet" href="style.css">
<div class="card" dir="rtl">
  <h4 style="color: #dc2626;">شروع یک نخ</h4>
  <p>
    برای اجرای نخ باید از متد <code>start()</code> کلاس Thread استفاده کنیم.  
    وقتی <code>start()</code> رو صدا می‌زنیم:
  </p>
  <ul>
    <li>یک نخ جدید با <em>call stack</em> جدا ساخته میشه.</li>
    <li>نخ از وضعیت New به وضعیت Runnable میره.</li>
    <li>وقتی Scheduler فرصت بده، متد <code>run()</code> اجرا میشه.</li>
  </ul>
</div>

```java
class MyRunnable implements Runnable {
    public void run() {
        System.out.println("Thread is running...");
    }

    public static void main(String[] args) {
        MyRunnable m1 = new MyRunnable();
        Thread t1 = new Thread(m1); // پاس دادن شیء Runnable به Thread
        t1.start(); // شروع نخ
    }
}

MyRunnable.main(new String[]{});
```

<div class="card" dir="rtl">
  <h4 style="color: #dc2626;">نکته مهم</h4>
  <ul>
    <li>اگه فقط کلاس رو بسازیم که Runnable رو پیاده‌سازی کرده باشه، اون شیء به خودی خود نخ حساب نمیشه.</li>
    <li>برای همین باید حتماً یک شیء از کلاس <code>Thread</code> بسازیم و شیء Runnable رو بهش پاس بدیم.</li>
    <li>با این کار <code>run()</code> داخل نخ جدید اجرا میشه.</li>
  </ul>
</div>

  <h4 style="color: #dc2626;">توضیح دقیق تفاوت run() و start()</h4>
  <p>
    متد <code>run()</code> در واقع فقط یک متد معمولی جاواست، مثل هر متد دیگه‌ای که در کلاس‌هاتون می‌نویسید.  
    وقتی شما مستقیم <code>run()</code> رو صدا می‌زنید، کدهای داخلش در همون نخ فعلی (مثلا <strong>main thread</strong>) اجرا میشن.  
    بنابراین هیچ نخ جدیدی ساخته نمیشه، هیچ <em>call stack</em> تازه‌ای ایجاد نمیشه و مدیریت حافظه هم فقط در همون نخ اصلی انجام میشه.
  </p>

  <p>
    اما وقتی متد <code>start()</code> رو صدا می‌زنید، اتفاقات زیر رخ میده:
  </p>
  <ul>
    <li>JVM یک <strong>Thread object</strong> واقعی ایجاد می‌کنه.</li>
    <li>برای اون نخ جدید یک <strong>call stack جدا</strong> در حافظه رزرو میشه (یعنی فضای مخصوص خودش برای اجرای متدها داره).</li>
    <li>Thread Scheduler تصمیم می‌گیره چه زمانی این نخ اجرا بشه. وقتی نوبتش برسه، متد <code>run()</code> درون اون نخ جدید اجرا میشه.</li>
    <li>به خاطر وجود <strong>call stack جدا</strong>، چند نخ می‌تونن به صورت همزمان (concurrent) اجرا بشن.</li>
  </ul>
  <div class = "images">
    <img src="img/pic3.png"/>
  </div>

  <h4 style="color: #dc2626;">چرا این تفاوت مهمه؟</h4>
  <p>
    فرض کنید شما می‌خواید دو کار رو همزمان انجام بدید، مثلا:
  </p>
  <ul>
    <li>نخ اول: دانلود فایل از اینترنت.</li>
    <li>نخ دوم: پردازش داده و نمایش روی صفحه.</li>
  </ul>
  <p>
    اگه از <code>run()</code> استفاده کنید، هر دو کار پشت سر هم و روی همون main thread اجرا میشن.  
    یعنی اولی باید تموم بشه تا دومی شروع بشه. در عمل <strong>هیچ مالتی‌تردینگی رخ نداده</strong>!
  </p>
  <p>
    اما اگه از <code>start()</code> استفاده کنید، برای هر کدوم یک نخ واقعی ساخته میشه با call stack جدا.  
    پس JVM می‌تونه بین اون‌ها جابه‌جا بشه و هر دو کار به صورت همزمان (یا دست‌کم به صورت شبیه‌سازی همزمان) پیش برن.  
    نتیجه‌اش اینه که برنامه سریع‌تر، روان‌تر و بهینه‌تر کار می‌کنه.
  </p>
</div>

```java
class DemoThread extends Thread {
    public void run() {
        System.out.println("Thread is running by: " + Thread.currentThread().getName());
    }

    public static void main(String[] args) {
        DemoThread t1 = new DemoThread();

        
        t1.run();    

        
        t1.start(); 
    }
}

DemoThread.main(new String[]{});
```

  <h2>Thread Scheduler در جاوا</h2>
  <p>
    وقتی چند نخ (Thread) در وضعیت <strong>Runnable</strong> هستن، این که دقیقا کدوم نخ اجرا بشه دست ما نیست!  
    تصمیم‌گیری در مورد اینکه چه زمانی و کدوم نخ باید اجرا بشه به عهده‌ی <strong>Thread Scheduler</strong> هست که بخشی از JVM محسوب میشه.
  </p>

  <h4 style="color: #dc2626;">ویژگی‌های Thread Scheduler</h4>
  <ul>
    <li>فقط یک نخ می‌تونه در یک لحظه روی یک پردازنده (CPU core) اجرا بشه.</li>
    <li>Thread Scheduler مشخص می‌کنه که در هر لحظه کدوم نخ اجرا بشه.</li>
    <li>هیچ تضمینی وجود نداره که نخ‌ها به ترتیب شروع‌شدن اجرا بشن.</li>
    <li>الگوریتم دقیق Scheduler به پیاده‌سازی JVM و سیستم‌عامل بستگی داره.</li>
  </ul>

  <h4 style="color: #dc2626;">نکته مهم</h4>
  <p>
    حتی اگر شما چند نخ بسازید و به ترتیب <code>start()</code> صدا بزنید، خروجی همیشه همون ترتیبی که انتظار دارید نخواهد بود.  
    این دقیقا به خاطر وجود Scheduler هست. پیشنهاد می‌شود کد زیر را در محیط Intellij Idea اجرا کنید.
  </p>
</div>

```java
class MyThread extends Thread {
    public void run() {
        for (int i = 1; i <= 3; i++) {
            System.out.println(Thread.currentThread().getName() + " is running... step " + i);
            try {
                Thread.sleep(500); 
            } catch (InterruptedException e) {
                e.printStackTrace();
            }
        }
    }

    public static void main(String[] args) {
        MyThread t1 = new MyThread();
        MyThread t2 = new MyThread();
        MyThread t3 = new MyThread();

        t1.start();
        t2.start();
        t3.start();
    }
}

MyThread.main(new String[]{});
```

  <h2>آیا می‌توان یک نخ را دوبار استارت کرد؟</h2>

  <h4 style="color: #dc2626;">پاسخ کوتاه</h4>
  <p>
    خیر ❌  
    وقتی یک نخ با <code>start()</code> آغاز میشه، دیگه نمی‌تونیم دوباره اون رو استارت کنیم.  
    اگر این کار رو بکنیم، یک خطای <code>IllegalThreadStateException</code> رخ میده.
  </p>

  <h4 style="color: #dc2626;">چرا؟</h4>
  <p>
    چون وقتی نخ برای بار اول استارت میشه، JVM یک <strong>call stack جدید</strong> برای اون نخ ایجاد می‌کنه.  
    بعد از اینکه نخ به وضعیت <strong>Terminated</strong> برسه، دیگه نمی‌تونیم همون نخ رو زنده کنیم.  
    در واقع هر نخ فقط یک بار می‌تونه از New به Runnable بره.
  </p>

  <h4 style="color: #dc2626;">مثال</h4>
</div>

```java
class MyThread extends Thread {
    public void run() {
        System.out.println("Thread is running...");
    }

    public static void main(String[] args) {
        MyThread t1 = new MyThread();
        t1.start(); // نخ برای اولین بار اجرا میشه

        // تلاش برای اجرای دوباره همون نخ
        t1.start(); // خطا: IllegalThreadStateException
    }
}

MyThread.main(new String[]{});
```

  <h2>متد join() در جاوا</h2>

  <h4 style="color: #dc2626;">تعریف</h4>
  <p>
    متد <code>join()</code> باعث میشه نخ فعلی منتظر بمونه تا نخ دیگری که روی اون صدا زده شده کارش رو تموم کنه.  
    به بیان ساده: اگه نخ A روی نخ B متد <code>join()</code> رو صدا بزنه، نخ A متوقف میشه تا B کامل اجرا بشه.
  </p>

  <h4 style="color: #dc2626;">سینتکس</h4>
  <ul>
    <li><code>public final void join() throws InterruptedException</code></li>
    <li><code>public final void join(long millis) throws InterruptedException</code></li>
    <li><code>public final void join(long millis, int nanos) throws InterruptedException</code></li>
  </ul>

  <h4 style="color: #dc2626;">مثال ساده</h4>
</div>

```java
class JoinExample extends Thread {
    public void run() {
        for (int i = 1; i <= 3; i++) {
            System.out.println(Thread.currentThread().getName() + " - Step " + i);
            try {
                Thread.sleep(500);
            } catch (InterruptedException e) {
                System.out.println(e);
            }
        }
    }

    public static void main(String[] args) {
        JoinExample t1 = new JoinExample();
        JoinExample t2 = new JoinExample();
        JoinExample t3 = new JoinExample();

        t1.start();
        try {
            t1.join();
        } catch (InterruptedException e) {
            System.out.println(e);
        }

        t2.start();
        t3.start();
    }
}

JoinExample.main(new String[]{});
```

  <h2>اولویت‌بندی نخ‌ها (Thread Priority)</h2>

  <h4 style="color: #dc2626;">تعریف</h4>
  <p>
    هر نخ در جاوا یک <strong>اولویت (Priority)</strong> داره که عددی بین 1 تا 10 هست.  
    به صورت پیش‌فرض همه نخ‌ها اولویت 5 دارن (<code>Thread.NORM_PRIORITY</code>).  
    این عدد در واقع فقط یک <strong>پیشنهاد</strong> به Thread Scheduler هست که نخ مهم‌تر رو در نظر بگیره.
  </p>

  <h4 style="color: #dc2626;">نکته مهم</h4>
  <p>
    اولویت نخ به تنهایی تعیین‌کننده نیست.  
    حتی اگه شما یک نخ رو با <code>MAX_PRIORITY</code> بسازید، هیچ تضمینی وجود نداره که همیشه زودتر از نخ‌های دیگه اجرا بشه.  
    چون اجرای واقعی نخ‌ها به <strong>Thread Scheduler</strong> وابسته‌ست و Scheduler هم توسط JVM و سیستم‌عامل مدیریت میشه.
  </p>

  <p>
    بنابراین:
  </p>
  <ul>
    <li>نخ با Priority بالاتر ممکنه دیرتر از نخ با Priority پایین‌تر اجرا بشه.</li>
    <li>رفتار دقیق نخ‌ها روی سیستم‌عامل‌های مختلف (مثل ویندوز، لینوکس یا مک) ممکنه متفاوت باشه.</li>
    <li>از Priority بیشتر برای «پیشنهاد دادن» استفاده میشه، نه برای کنترل قطعی ترتیب اجرای نخ‌ها.</li>
  </ul>

   <h4 style="color: #dc2626;">مقدارهای ثابت پیش‌فرض</h4>
  <ul>
    <li><code>Thread.MIN_PRIORITY = 1</code></li>
    <li><code>Thread.NORM_PRIORITY = 5</code> (پیش‌فرض همه نخ‌ها)</li>
    <li><code>Thread.MAX_PRIORITY = 10</code></li>
  </ul>

  <h4 style="color: #dc2626;">متدهای مربوط</h4>
  <ul>
    <li><code>public final int getPriority()</code> → دریافت اولویت نخ</li>
    <li><code>public final void setPriority(int priority)</code> → تغییر اولویت نخ</li>
  </ul>

  <h4 style="color: #dc2626;">مثال</h4>
</div>

```java
class PriorityExample extends Thread {
    public void run() {
        System.out.println("Running thread: " 
            + Thread.currentThread().getName() 
            + " with priority: " 
            + Thread.currentThread().getPriority());
    }

    public static void main(String[] args) {
        PriorityExample t1 = new PriorityExample();
        PriorityExample t2 = new PriorityExample();
        PriorityExample t3 = new PriorityExample();

        t1.setName("LowPriorityThread");
        t2.setName("NormalPriorityThread");
        t3.setName("HighPriorityThread");

        t1.setPriority(Thread.MIN_PRIORITY);   // 1
        t2.setPriority(Thread.NORM_PRIORITY);  // 5
        t3.setPriority(Thread.MAX_PRIORITY);   // 10

        t1.start();
        t2.start();
        t3.start();
    }
}

PriorityExample.main(new String[]{});
```

  <h2>مطالعه بیشتر: الگوریتم‌های زمان‌بندی در Scheduler</h2>

  <h4 style="color: #dc2626;">First Come First Serve (FCFS)</h4>
  <p>
    ساده‌ترین الگوریتم زمان‌بندی.  
    نخ‌ها به ترتیبی که درخواست اجرا شدن وارد صف میشن، همون ترتیب هم توسط CPU اجرا میشن.  
    مثل صف نانوایی: هر کی زودتر بیاد، زودتر نوبتش میشه.
  </p>

  <h4 style="color: #dc2626;">Shortest Job Next (SJN) / Shortest Job First (SJF)</h4>
  <p>
    در این الگوریتم، نخ‌هایی که کار کمتری دارن (یعنی زمان اجرای کمتری می‌برن) زودتر انتخاب میشن.  
    مشکلش اینه که نخ‌های طولانی ممکنه مدت زیادی پشت صف بمونن (Starvation).
  </p>

  <h4 style="color: #dc2626;">Round Robin (RR)</h4>
  <p> 
    هر نخ یک بازه زمانی مشخص (Time Slice یا Quantum) داره.  
    وقتی زمانش تموم شد، CPU نخ بعدی رو انتخاب می‌کنه.  
    این چرخه همینطور ادامه پیدا می‌کنه تا همه نخ‌ها فرصت اجرا داشته باشن.  
    این روش خیلی شبیه به <strong>context switching</strong> در سیستم‌های واقعی هست.
  </p>

  <h4 style="color: #dc2626;">Priority Scheduling</h4>
  <p>
    در این روش هر نخ یک <strong>اولویت (Priority)</strong> داره و Scheduler نخ‌ها رو براساس اون انتخاب می‌کنه.  
    نخ‌هایی با Priority بالاتر احتمال بیشتری دارن که زودتر اجرا بشن.  
    اما یک نکته مهم: همیشه هم نخ با Priority بالاتر اجرا نمیشه، چون در عمل Scheduler باید تعادل بین نخ‌ها رو حفظ کنه.
  </p>

  <h4 style="color: #dc2626;">مثال: الگوریتم لاتاری (Lottery Scheduling)</h4>
  <p>
    یکی از پیاده‌سازی‌های جالب Priority Scheduling الگوریتم لاتاری هست.  
    توی این روش، هر نخ براساس Priority که داره یک تعداد بلیت لاتاری دریافت می‌کنه.  
    Scheduler در هر بار انتخاب، به صورت تصادفی یک بلیت رو برمی‌داره و نخ صاحب اون بلیت اجرا میشه.
  </p>
  <p>
    برای مثال:
  </p>
  <ul>
  <li>نخ A با Priority=3 یعنی 3 بلیت دریافت می‌کند.</li>
  <li>نخ B با Priority=1 یعنی 1 بلیت دریافت می‌کند.</li>
  <li>نخ C با Priority=6 یعنی 6 بلیت دریافت می‌کند.</li>
  </ul>
  <p>
    در این حالت احتمال انتخاب نخ C بیشتر از بقیه‌ست، اما هنوز هم امکان داره نخ A یا B انتخاب بشن.  
    این روش باعث میشه هم نخ‌های مهم‌تر شانس بیشتری داشته باشن و هم نخ‌های کم‌اهمیت کامل کنار گذاشته نشن.
  </p>
